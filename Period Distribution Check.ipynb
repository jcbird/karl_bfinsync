```python
import numpy as np
import matplotlib as mpl
mpl.use('Agg')
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
periodx = 10**(np.random.normal(loc=4.8, scale=2.3, size=1000000))
```

```python
plt.figure(figsize=(8,6))
plt.hist(np.log10(periodx), bins=100)
plt.axvline(4.8, c='red')
plt.xlabel('log10(Period)')
plt.ylabel('N')
```
